```python
# Line Plot
x = [12,16,7,21,8]
y = [34,23,9,11,5]
plt.plot(x,y, color = 'darkred', linestyle = '--', marker='>', label='datapoints')
plt.xlabel('Student')
plt.ylabel('Age')
plt.title('College')
plt.legend()
plt.grid(axis = 'y')
plt.show()
```

```python
# Scatter Plot
x = [12,16,7,21,8]
y = [24,32,14,42,16]
plt.scatter(x,y, color = 'red', marker= 'v', label= 'y=x*2')
plt.xlabel('Student')
plt.ylabel('Age')
plt.title('College')
plt.legend()
plt.grid(axis = 'y')
plt.show()
```

```python
x = [12,16,7,21,8]
y = [24,32,14,42,16]
sizes = [32,45,10,35,46]
plt.scatter(x,y, color = 'red', s=sizes, marker='v', label='y=x**2')
plt.title('College')
plt.xlabel('Student')
plt.ylabel('Age')
plt.legend()
plt.grid(axis = 'y')
plt.show()
```

```python
sizes = [40,20,50,10]
label = ['grp 1','grp 2','grp 3','grp 4']
color = ['red','green','yellow','blue']
plt.pie(sizes, labels=label, autopct='%1.1f%%',startangle=90, colors=color)
plt.title('Pie Graph')
plt.show()
```

```python
sizes = [40,20,50,10]
label = ['grp 1','grp 2','grp 3','grp 4']
color = ['blue','yellow','green','red']
plt.pie(sizes, colors=color, labels=label, autopct='%1.1f%%', startangle=90, explode=(0.1,0,0.1,0))
plt.title('Pie Graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
w = 0.2
x = ['C1','C2','C3','C4']
boys = [100,40,60,50]
girls = [200,70,30,70]
bar1 = np.arange (len(x))
plt.bar(bar1,boys,w, label= 'boys')
bar2 = [i+w for i in bar1]
plt.bar(bar2,girls,w,label= 'girls')
plt.xticks(bar1+w/2,x)
plt.show()
```

```python
w = 0.2
x = ['C1','C2','C3','C4']
boys = [100,40,60,50]
girls = [200,50,60,90]
both = [10,60,30,80]
bar1=np.arange(len(x))
plt.bar(bar1,boys,w,label='boyes')
bar2= [i+w for i in bar1]
plt.bar(bar2,girls,w,label='girls')
bar3= [i+2*w for i in bar1]
plt.bar(bar3,both,w,label='both')
plt.xticks(bar1+w/2,x)
plt.show()
```
